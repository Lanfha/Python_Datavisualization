```python
import pandas as pd
```

## Dictionary / Row Oriented

```python
sales1 = [{'account':'Jones LLC','Jan':'150','Feb':'200','Mar':'140'},
              {'account':'Alpha Co','Jan':'200','Feb':'210','Mar':'215'},
              {'account':'Blue Inc','Jan':'50','Feb':'90','Mar':'95'}]
df1 = pd.DataFrame(sales1)
df1
```

## List Row Oriented
### จากรูปที่โหลดมาผิดตรง List Row Oriented
มีการเรียงตัวเลขผิด โดยแก้ใหม่คือ สลับ 50 กับ 140 และ90กับ 215 จึงจะถูกต้อง

```python
sales2 = [('Jones LLC','150','200','140'),
              ('Alpha Co','200','210','215'),
              ('Blue Inc','50','90','95')]
labels = ['account','Jan','Feb','Mar']
df2 = pd.DataFrame.from_records(sales2,columns = labels)
df2
```

## Dictionary Cotumn Oriented

```python
sales3 = {'account': ['Jones LLC', 'Alpha co', 'Blue Inc'], 
             'Jan': [150, 200, 50], 
             'Feb': [200, 210, 90], 
             'Mar': [140, 215, 95]}
df3 = pd.DataFrame.from_dict(sales3)
df3
```

## List Contumn Oriented

```python
sales4 = [('account', ['Jones LLC', 'Alpha co', 'Blue Inc']), 
            ('Jan', [150, 200, 50]), 
            ('Feb', [200, 210, 90]), 
            ('Mar', [140, 215, 95]) ]
df4 = pd. DataFrame.from_items (sales4)
df4
```
