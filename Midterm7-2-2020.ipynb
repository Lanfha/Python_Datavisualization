1> สร้าง Function ชื่อ smartDict ซึ่งทำหน้าที่:
    -Function รับ input เป็น list 2 ตัวที่มีขนาดเท่ากันโดย list แรกเป็น list ของตัวเลข list สองเป็น list ของ text
    -ถ้าค่าใน list แรกเป็นเเลข คู่ ให้ค่าใน dict ที่เป็น output มี key เป็นค่าจากlistตัวที่ 1 และvalueจาก list ตัวที่ 2
    -ถ้าค่าใน list แรกเป็นเเลข คี่ ให้ค่าใน dict ที่เป็น output มี key เป็นค่าจากlistตัวที่ 2 และvalueจาก list ตัวที่ 1
    -output ของfuntion คือ dict ที่สร้างขึ้น

```python
import pandas as pd
```

```python
def smartDict (listnumber,listtext):
    f ={}
    for i in range(len(listnumber)):
        if listnumber[i]%2==1:
            f[listnumber[i]] = listtext[i]
        else:
            f[listtext[i]] = listnumber[i]
    return (f)    
```

```python
output1 = smartDict([1,2,3],['apple','banana','orange'])
```

2> จาหไฟล์ Wuhan_nCoV_GNome.text จงเขียนโปรแกรมเพื่อตอบคำถามต่อไปนี้                             
-กรดอะมิโนใด(A,T,G,C)มีปริมาณมากที่สุดและมีเท่าไหร่

```python
DNAseq = pd.read_csv('C:\\Users\HP\OneDrive\Desktop\Wuhan_nCoV_GNome.txt', header=None)
DNAseq.head()
```

```python
def countChar(seq,character):
    charcount = 0
    for char in seq:
        if char == character:
            charcount = charcount + 1
    return charcount  
```

```python
ANum = 0
TNum = 0
GNum = 0
CNum = 0
for row in range(DNAseq.shape[0]):
    ANum += countChar(DNAseq.iloc[row][0],'A')
    TNum += countChar(DNAseq.iloc[row][0],'T')
    GNum += countChar(DNAseq.iloc[row][0],'G')
    CNum += countChar(DNAseq.iloc[row][0],'C')
    
print(f'A มีปริมาณ {ANum} ตัว,T มีปริมาณ {TNum} ตัว,G มีปริมาณ{GNum} ตัว,C มีปริมาณ{CNum} ตัว')
```

-จงหา sequence ที่มีขนาด 3 และ 4 ที่เกิดขึ้นซ้ำมากที่สุด

```python
def populate3charstring(seq):
    str_list = list()
    for i in range(len(seq)-2):
        str_list.append(seq[i:i+3])
    return str_list    
```

```python
def populate4charstring(seq):
    str_list = list()
    for i in range(len(seq)-3):
        str_list.append(seq[i:i+4])
    return str_list    
```

```python
populate4charstring('CGAACTTTAAAT')
```

```python
populate3charstring('CGAACTTTAAAT')
```

```python
list_all_4char = list()
for row in range(DNAseq.shape[0]):
    list_all_4char = list_all_4char + populate4charstring(DNAseq.iloc[row][0])
```

```python
list_all_3char = list()
for row in range(DNAseq.shape[0]):
    list_all_3char = list_all_3char + populate3charstring(DNAseq.iloc[row][0])
```
