```python
list_word = []
for i in range(Data.shape[0]) :
    A = Data.iloc[i]['ภาษาอีสาน'].split('/')
    B = Data.iloc[i]['ภาษาไทย'].split('/')
    if len(A) == len(B):
        X = len(A)
        for j in range(X):
            list_word.append(f'ภาษาอีสานคำว่า "{A[j]}" มีความหมายว่า "{B[j]}"')
```
