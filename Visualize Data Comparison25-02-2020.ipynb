### Barchat

## สร้างกราฟแท่งของรายได้ในปี 2541 ของจังหวัดกรุงเทพและปริมณฑล

```python
plt.bar([1,2,3,4],greaterbangkok.iloc[:,2])
plt.title('Barchart compareing incomes of provinced in Greater Bangkok in 2541')
plt.xlabel('Provinces')
plt.ylabel('Incom Baht/Month')
```

### เพิ่มชื่อในแต่ละแกน tick_label = ชื่อไฟล์.iloc[แถวที่เลือก,คอลัมน์]

```python
plt.bar([1,2,3,4],greaterbangkok.iloc[:,2],color = 'Pink',tick_label=greaterbangkok.iloc[:,-1])
plt.title('Barchart compareing incomes of provinced in Greater Bangkok in 2541')
plt.xlabel('Provinces')
plt.ylabel('Incom Baht/Month/Household')
```

### compare different years

```python
plt.bar([1-(0.35/2),2-(0.35/2),3-(0.35/2),4-(0.35/2)],greaterbangkok.iloc[:,-2],color = 'Pink',width=0.35,tick_label=greaterbangkok.iloc[:,-1],label='2558',align='edge')
plt.bar([1+(0.35/2),2+(0.35/2),3+(0.35/2),4+(0.35/2)],greaterbangkok.iloc[:,2],color = 'Y',width=0.35,tick_label=greaterbangkok.iloc[:,-1],label='2541',align='edge')
plt.title('Barchart compareing incomes of provinced in Greater Bangkok in 2541,2558')
plt.xlabel('Provinces')
plt.legend()
plt.ylabel('Incom Baht/Month/Household')
```

### Stacked Barchart

```python
saledata = pd.DataFrame([{'North':20,'East':10,'West':90,'South':120},
                        {'North':220,'East':20,'West':50,'South':90}])
saledata
```

```python
plt.bar([1,2],saledata.iloc[:,0],tick_label=['Company A','Company B'])
plt.bar([1,2],saledata.iloc[:,1],bottom=saledata.iloc[:,0],tick_label=['Company A','Company B'])
plt.bar([1,2],saledata.iloc[:,2],bottom=saledata.iloc[:,0]+saledata.iloc[:,1],tick_label=['Company A','Company B'])
plt.bar([1,2],saledata.iloc[:,3],bottom=saledata.iloc[:,0]+saledata.iloc[:,1]++saledata.iloc[:,2],tick_label=['Company A','Company B'])
```
