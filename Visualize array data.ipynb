```python
import pandas as pd
```

```python
irisdata = pd.read_csv('D:\\iris.data',header=None)
irisdata
```

```python
import numpy as np
```

```python
data_np = irisdata.iloc[:,:-1].to_numpy()
data_np
```

```python
setosa_np = irisdata.iloc[:50,:-1].to_numpy()
setosa_np                    
```

```python
versicolor_np = irisdata.iloc[50:100,:-1].to_numpy()
virginica_np = irisdata.iloc[100:,:-1].to_numpy()
```

## Visualize array

```python
from matplotlib import pyplot as plt
import matplotlib
```

### สีสว่างคือค่ามาก

```python
matplotlib.rcParams['figure.figsize'] = [12,12]
plt.imshow(setosa_np, interpolation='nearest')
plt.show()
```

```python
matplotlib.rcParams['figure.figsize'] = [12,12]
plt.imshow(versicolor_np, interpolation='nearest')
plt.show()
```

```python
matplotlib.rcParams['figure.figsize'] = [12,12]
plt.imshow(virginica_np, interpolation='nearest')
plt.show()
```

```python
matplotlib.rcParams['figure.figsize'] = [12,12]
plt.subplot(1, 3, 1)
plt.imshow(setosa_np, interpolation='nearest')
plt.subplot(1, 3, 2)
plt.imshow(versicolor_np, interpolation='nearest')
plt.subplot(1, 3, 3)
plt.imshow(virginica_np, interpolation='nearest')
plt.show()
```

```python
matplotlib.rcParams['figure.figsize'] = [40,40]
plt.imshow(data_np, interpolation='nearest')
plt.show()
```
